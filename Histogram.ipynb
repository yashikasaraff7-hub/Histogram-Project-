```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ages = np.random.normal(18,20,100)
plt.hist(ages, bins=10, edgecolor="black",color="purple")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.title("Data Set")
plt.show()
```

```python
df = pd.read_csv("./students.csv")
number_of_siblings = df["Number of sisters/brothers"]

plt.hist(number_of_siblings, bins=10, edgecolor="black",color="yellow")
plt.xlabel("Number of siblings")
plt.ylabel("Number of Students")
plt.title("Data")
plt.show()
```
